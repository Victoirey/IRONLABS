# Challenge 1 - Generate and Plot Normal Distributions
#### Step 1: Generate samples and test normal distribution.

Use mean=50, standard_deviation=5, and sample_size=[10, 50, 500, 5000] to generate 4 random samples that are normally distributed.

**Hint**: Read the documentation for `scipy.stats.norm.rvs`. The function will help you create the normal random samples.

```python
# your code here
from scipy.stats import norm
mean=50
standard_deviation=5
sample_size=[10, 50, 500, 5000]
samples=[]
for i in sample_size:
    r = norm.rvs(loc=mean, scale=standard_deviation, size=i)
    samples.append(r)
print('Sample 1: ', samples[0])
print('Sample 2: ', samples[1])
print('Sample 3: ', samples[2])
print('Sample 4: ', samples[3])
```

```python
print('Sample 1: ', len(samples[0]))
print('Sample 2: ', len(samples[1]))
print('Sample 3: ', len(samples[2]))
print('Sample 4: ', len(samples[3]))
```

# Challenge 4 - Central Limit Theorem

A delivery company needs 35 minutes to deliver a package, with a standard deviation of 8 minutes. Suppose that in one day, they deliver 200 packages.

**Hint**: `stats.norm.cdf` can help you find the answers.

#### Step 1: What is the probability that the mean delivery time today is between 30 and 35 minutes?

```python
# your code here
# for total population, we have 8min std dev and mean 35
m=35 
#std error is calculated for my sample, stdev taken into account the number of elements in my sample
#8 is the std dev of my population but it should be a diff stdev for my sample
se=8/np.sqrt(200)
cdf_probability = stats.norm.cdf(35, m, se) - stats.norm.cdf(30, m, se)
cdf_probability
```

```python
m=35 
#std error is calculated for my sample 
stdv=8
cdf_probability = stats.norm.cdf(35, m, stdv) - stats.norm.cdf(30, m, stdv)
cdf_probability
```

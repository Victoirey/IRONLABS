### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(moovies.director_name, moovies.country, margins=True, margins_name="Total")
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab(moovies.director_name, [moovies.country, moovies.genres], margins=True, margins_name="Total")
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(moovies.genres, moovies.color, values=moovies.duration, aggfunc='mean', margins=True, margins_name="Total").fillna('-')
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(moovies.genres, moovies.color, values=moovies.duration, aggfunc='mean', margins=True, margins_name="Total").fillna('-')
```

```python
moovies['film_likes']=moovies['actor_3_facebook_likes']+moovies['actor_1_facebook_likes']+moovies['actor_2_facebook_likes']
```

```python
moovies['director_likes']=moovies['cast_total_facebook_likes']-moovies['film_likes']
```

```python
directors=pd.crosstab(moovies.director_name, columns='Likes', values=moovies.director_likes, aggfunc='mean',margins=True, margins_name="Total").round(0)
directors
```

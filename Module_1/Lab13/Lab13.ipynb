```python
import requests
req_2 = requests.get('https://api.github.com/search/code?q=.scavengerhunt+in:path+repo:ironhack-datalabs/scavenger')
search = req_2.json()
search
```

```python
files = search['items'];
files = sorted(files, key=lambda file: file['name']);
files
```

```python
contents_list = []
headers = {'Accept': 'application/vnd.github.v3.raw', };
headers
```

```python
#get content per path and combine in one list
for file in files:
    content_path = ('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}').format(file['path']);
    response = requests.get(content_path, headers=headers);
    contents_list.append(response.text)
contents_list
```

```python
clean_contents = [word.strip('\n') for word in contents_list]
joke = " ".join(clean_contents)
```

```python
joke = ''
for word in joke:
    answer_bytes = word.encode('ascii')
    messegae_bytes =base64.b64decode(answer_bytes)
    message = messegae_bytes.decode('ascii')
    joke += message.strip('\n') + ' '
#print(message)
print(joke)
```

```python
#another way
repo_files_names = [item["name"] for item in search["items"]]
repo_files_names.sort()

repo_files_urls = {item["name"]: item["url"] for item in search["items"]}
repo_files_urls
```

```python
repo_files = [requests.get(repo_files_urls[repo_filename]) for repo_filename in repo_files_names]
repo_files
```

```python
repo_files_content = [file.json()["content"] for file in repo_files]
repo_files_content
```

```python
import base64

message = [base64.b64decode(b64message).decode('ascii') for b64message in repo_files_content]
message
```

```python
joke = ''.join(message).replace('\n', ' ')
joke
```

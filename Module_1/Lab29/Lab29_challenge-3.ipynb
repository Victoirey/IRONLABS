```python
customer_spent = orders.groupby('CustomerID', as_index=False).agg({'amount_spent':'sum'})
```

```python
customer_spent
```

```python
vip_customer = customer_spent.loc[customer_spent['amount_spent'] >= customer_spent.amount_spent.quantile(0.95)]
preferred_customer = customer_spent.loc[(customer_spent['amount_spent'] >= customer_spent.amount_spent.quantile(0.75)) & (customer_spent['amount_spent'] < customer_spent.amount_spent.quantile(0.95))]
```

```python
customer_spent
```

```python
vip_customer
```

```python
preferred_customer
```

```python
sum_cust=orders['amount_spent'].groupby(orders['CustomerID']).sum()
```

```python
sum_cust
```

```python
sum_cust[17850]
```

```python
sum_cust.columns = ['CustomerID', 'Sum/Cust']
```

```python
sum_cust
```

```python
type(sum_cust)
```

```python
orders = pd.merge(orders, sum_cust, on='CustomerID')
```

```python
orders
```

```python
orders = orders.rename(columns={'amount_spent_y':'Amount/Customer'})
orders
```

```python
np.quantile(orders['Amount/Customer'], 0.95)
```

```python
np.quantile(orders['Amount/Customer'], 0.75)
```

```python
orders['Client type']=np.where((orders['Amount/Customer']>=np.quantile(orders['Amount/Customer'], 0.95)), 'VIP', np.where((orders['Amount/Customer']>=np.quantile(orders['Amount/Customer'], 0.75), 'Preferred Customers', 'Normal Customers')))
```

## Q2: How to identify which country has the most VIP Customers?

```python
vip_countries=pd.merge(vip_customer, orders, on='CustomerID')
```

```python
vip_countries=vip_countries.groupby('Country',as_index=False).agg({'CustomerID':'nunique'})
```

```python
vip_countries.sort_values('CustomerID', ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
preferred_countries=pd.merge(preferred_customer, orders, on='CustomerID')
```

```python
preferred_countries=preferred_countries.groupby('Country', as_index=False).agg({'CustomerID':'nunique'})
```

```python
preferred_countries.sort_values('Country', ascending=False)
```

```python
combined_countries=pd.merge(vip_countries, preferred_countries, on='Country')
```

```python
combined_countries.sort_values('Country', ascending=False)
```
